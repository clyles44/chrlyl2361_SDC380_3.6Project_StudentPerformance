```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Excel file using correct path
df = pd.read_excel(r"C:\Files\Excel\Project\StudentsPerformance.xlsx")

df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
plt.figure()
plt.hist(df['math score'])

plt.title("Histogram of Math Scores\nStudentID: chrlyl2361 | Date: 03/31/2026")
plt.xlabel("Math Score")
plt.ylabel("Frequency")

plt.show()
```

```python
avg_scores = df.groupby('race/ethnicity')['math score'].mean()

plt.figure()
avg_scores.plot(kind='bar')

plt.title("Average Math Scores by Race/Ethnicity\nStudentID: chrlyl2361 | Date: 03/31/2026")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Average Math Score")

plt.show()
```

```python
plt.figure()
plt.scatter(df['math score'], df['writing score'])

plt.title("Math vs Writing Scores\nStudentID: chrlyl2361 | Date: 03/31/2026")
plt.xlabel("Math Score")
plt.ylabel("Writing Score")

plt.show()
```

```python
plt.figure()
plt.scatter(df['math score'], df['reading score'])

plt.title("Math Score vs Reading Score\nStudentID: chrlyl2361 | Date: 03/31/2026")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")

plt.show()
```

```python
plt.figure()
corr = df[['math score', 'reading score', 'writing score']].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap\nStudentID: chrlyl2361 | Date: 03/31/2026")

plt.show()
```
